```python
from langchain.tools import BaseTool, StructuredTool, tool
import googlemaps
def calculate_distance_and_time(origin: str, destination: str, mode: str = 'driving') -> dict:
    """
    Calculate the distance and estimated travel time between two locations using the Google Maps Distance Matrix API.

    Parameters:
    - origin (str): The starting location's address or coordinates.
    - destination (str): The destination location's address or coordinates.
    - mode (str, optional): The mode of transportation (e.g., 'driving', 'walking', 'bicycling', 'transit'). Default is 'driving'.

    Returns:
    - str: The distance and duration between the origin and destination in human-readable format.
    
    Example:
    distance, duration = calculate_distance_and_time("New York, NY", "Los Angeles, CA", mode='driving')
    """
    
    # Initialize the Google Maps client with your API key
    gmaps = googlemaps.Client(key="")
    
    # Request distance and time information using the Distance Matrix API
    result = gmaps.distance_matrix(origins=origin, 
                                   destinations=destination, 
                                   mode=mode)
    
    # Extracting the relevant information from the result
    distance = result['rows'][0]['elements'][0]['distance']['value']
    duration = result['rows'][0]['elements'][0]['duration']['value']
    print(distance,duration)
    return {"distance" : distance, "duration": duration}

distance_calculator = StructuredTool.from_function(
    func=calculate_distance_and_time,
    name="distance_calculator",
    description="""
    Calculate the distance and estimated travel time between two locations using the Google Maps Distance Matrix API.

    Parameters:
    - origin (str): The starting location's address or coordinates.
    - destination (str): The destination location's address or coordinates.
    - mode (str, optional): The mode of transportation (e.g., 'driving', 'walking', 'bicycling', 'transit'). Default is 'driving'.

    Returns:
    - dict: The distance and duration between the origin and destination in dictionary format.
    
    Example:
    distance, duration = calculate_distance_and_time("New York, NY", "Los Angeles, CA", mode='driving')
    """,
    return_direct=True,
)
```

```python
from langchain_core.pydantic_v1 import BaseModel, Field, validator
from langchain_core.output_parsers import JsonOutputParser

def profile_matcher(question, profile):
    class Score(BaseModel):
        profile_id : str = Field(description="profile id of the pal")
        compatability_score : float = Field(description = "Overall match percentage between the tourist and the pal, indicating how well they align based on interests, skills, and location.",
            ge=0, le=100
        )
        reasons : str = Field(description = "Key reasons for match")
        pal_location: str = Field(description = "Address of pal from their profile")
        tourist_location : str = Field(description= "Adress of the tourist")
        distance :int = Field(description = "distance between Pal and Tourist")
        time: str = Field(description = "duration required for Pal to cover the distance")
    parser = JsonOutputParser(pydantic_object=Score)
    query = f"""You are a Matching Agent tasked with pairing tourists with local Pals based on their profiles. Given the tourist's and Pals' profiles, evaluate their compatibility and suggest potential matches. Focus on shared interests, personality compatibility, and travel preferences.

        **Tourist Profile:**
        {question}
        
        **Pal Profiles:**
        {profile}
        
        Key factors to consider (but not limited to):
        
        1. **Interest Match:** Prioritize locals whose interests align closely with the tourist's, using weighted scoring to highlight key areas like art, food, and outdoor activities. Consider subcategory matches for detailed alignment.
        
        2. **Language Skills:** Match based on fluency and cultural relevance, prioritizing languages most important to the tourist. Adjust scores according to fluency levels.
        
        3. **Specializations:** Focus on locals with expertise relevant to the tourist's specific interests, considering their experience level and ability to customize experiences.
        
        4. **Location Suitability:** Evaluate proximity and convenience, considering not just travel time but also neighborhood relevance to the tourist's preferences. Ensure that the distance between the tourist and local is within 5KM.
        
        5. **Overall Score:** Calculate a weighted average of all criteria, ensuring that top matches meet minimum thresholds for quality and relevance.


        format instructions:
        provide JSON data of how you recomended the place
        

        """
        
    llm = ChatVertexAI (model_name="gemini-1.5-flash", temperature=0, project="")
    agent = create_tool_calling_agent(llm, tools, prompt)
    agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

    matched = agent_executor.invoke({"input":query})
    return (matched)
```
